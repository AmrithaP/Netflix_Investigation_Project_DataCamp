Answer the questions 
1. What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration.?
2. A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.?




```python
# 1
import numpy as np

# Find movies that are released in the year 1990s
movie_1990s = netflix_df[np.logical_and(np.logical_and(netflix_df['release_year']>=1990,netflix_df['release_year']<=1999), netflix_df['type']=='Movie')]

movie_1990s
```

```python
movie_1990s['show_id'].count()
```

There are total 183 movies which were released in 1990s

```python
# sort the movies of 1990 by duration
movie_1990s = movie_1990s.sort_values('duration')
movie_1990s
```

```python
# Find the most frequent movie duration
most_frequent_duration = movie_1990s['duration'].mode()[0]
#most_frequent_duration

import matplotlib.pyplot as plt

plt.hist(movie_1990s['duration'], bins=10)
plt.show()
```

```python
# Save the result as an integer called duration
duration = int(most_frequent_duration)
print(duration)
```

From the observation we can see that the Most Frequent Movie duration in 1990s is **94**.

```python
# 2.
# Find action movies released in 1990s

action_movie_1990s = netflix_df[np.logical_and(np.logical_and(netflix_df['release_year']>=1990,netflix_df['release_year']<=1999), netflix_df['genre']=='Action', netflix_df['type']=='Movie')]

action_movie_1990s
```

```python
action_movie_1990s['show_id'].count()
```

There are 48 Action Movies released in the year 1990

```python
# Check the count of movies whose duration is less than 90 minutes i.e. short

short_movie_count = int(action_movie_1990s[action_movie_1990s['duration']<90]['show_id'].count())
short_movie_count
```

There are 7 short action movies released in the year 1990
